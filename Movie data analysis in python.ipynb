```python
# Import the required libraries for the analysis
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('movies.csv')
df
```

```python
# Check for duplicates
print(df.duplicated())
```

```python
# Check for 'NAN' OR NOT A NUMBER as a missing data
print(df.isnull(). values.sum() != 0)
```

```python
# Check for missing values in the data
df.isnull()
```

```python
# Count no of missing values
print(df.isnull(). sum())
```

```python
# Check for 'NAN' OR NOT A NUMBER as a missing data
print(df!= float('NAN'))
```

```python
# Drop NaN and replace with 0
df1 = df.replace('NaN', np.nan)
```

```python
# Replace 'NaN' with 0
df2 = df1.fillna(0)
```

```python
# Confirm missing values sorted
print(df2.isnull(). sum())
```

```python
# Drop unwanted columns
df3 = df2.drop(columns = ['writer', 'star', 'country'])
df3
```

```python
# Check missing spellings in the name column
df3['name'].unique()
```

```python
# Creat a new column 'net income' from budget and gross fields
df3 = df2.assign(net_income = lambda x: x.gross - x.budget)
df3.head()
```

```python
# Order by net income in desecending order
df3.sort_values(by = ['budget'], inplace = False, ascending = False)
```

```python
# Explore the cleaned data
print(df3.shape)
df3.describe()
```

```python
# Slicing genre column
df3['genre']
```

```python
# A barchart with sns showing count of movie genre in descending order
sns.countplot(y = 'genre',data = df3, order = df3['genre'].value_counts().index)
plt.figure(figsize=(14,10))
matplotlib.rcParams['figure.figsize'] = (14, 8)
plt.show()
```

```python
# Creat a correlation marix and heatmap using the new dataframe

cor = df3.corr(method = 'pearson')

# The heatmap shows gross and votes are strongly correlated with budget

sns.heatmap(cor, annot = True)
```

```python
# Display high correlated variables > 0.6
cor = df3.corr(method = 'pearson')
Pair = cor
corr_pair = Pair.unstack()
corr_sort = corr_pair.sort_values()
High_corr = corr_sort[(corr_sort) > 0.6]
High_corr
```

```python
# Scatter plot of the correlated variables (net_income vs budget)

# Add top 3 movies with highest net income

plt.scatter('budget', 'net_income', data = df3, alpha = 0.9, s = 5)
plt.title('Budget vs Net Income')
plt.xlabel('Budget')
plt.ylabel('Net Earnings')
plt.text(237000000.0, 2.610246e+09, 'Avatar')
plt.text(356000000.0, 2.441501e+09, 'Avengers: Endgame')
plt.text(200000000.0, 2.001647e+09, 'Titanic')
plt.xscale('log')
plt.show()
```

```python
# Scatter plot of the correlated variables (votes vs gross)

plt.scatter('votes', 'net_income', data = df3, alpha = 0.9, s = 5 )
plt.title('Votes received vs Net income')
plt.xlabel('Votes received')
plt.ylabel('Net Earnings')
plt.text(1100000.0, 2.610246e+09, 'Avatar')
plt.text(903000.0, 2.441501e+09, 'Avengers: Endgame')
plt.text(1100000.0, 2.001647e+09, 'Titanic')
plt.xscale('log')
plt.show()
```

```python
# Result shows that (1) net earnings is positively associated with cost of production and public perception of the movie
# (2) Avatar followed by Avengers:Endagame and Titanic have the highest earnings.
```
