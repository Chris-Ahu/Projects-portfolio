```python
# Replace 'NaN' with 0

df2 = df1.fillna(0)
df2
```

```python
# Confirm missing values sorted

print(df2.isnull(). sum())
```

```python
# Creat a new column 'net income' from budget and gross fields

df3 = df2.assign(net_income = lambda x: x.gross - x.budget)
```

```python
# Explore the top six rows of the new data-frame

df3.head()
```

```python
# # Order by net income in desecending order

df3.sort_values(by = ['net_income'], inplace = False, ascending = False)
```

```python
# Exploring the data

print(df3.shape)
df3.describe()
```

```python
# Creat a correlation marix and heatmap using the new dataframe

cor = df3.corr(method = 'pearson')

# The heatmap shows gross and votes are strongly correlated with budget

sns.heatmap(cor, annot = True)
```

```python
# Display high correlated variables > 0.6

cor = df3.corr(method = 'pearson')
Pair = cor
corr_pair = Pair.unstack()
corr_sort = corr_pair.sort_values()
High_corr = corr_sort[(corr_sort) > 0.6]
High_corr
```

```python
# Scatter plot of the correlated variables (net_income vs budget)

# Add top 3 movies with highest net income

plt.scatter('budget', 'net_income', data = df3, alpha = 0.9)
plt.title('Budget vs Net Income')
plt.xlabel('Budget')
plt.ylabel('Net Earnings')
plt.text(237000000.0, 2.610246e+09, 'Avatar')
plt.text(356000000.0, 2.441501e+09, 'Avengers: Endgame')
plt.text(200000000.0, 2.001647e+09, 'Titanic')
plt.xscale('log')
plt.show()
```

```python
# Scatter plot of the correlated variables (votes vs gross)

plt.scatter('votes', 'net_income', data = df3, alpha = 0.9 )
plt.title('Votes received vs Net income')
plt.xlabel('Votes received')
plt.ylabel('Net Earnings')
plt.text(1100000.0, 2.610246e+09, 'Avatar')
plt.text(903000.0, 2.441501e+09, 'Avengers: Endgame')
plt.text(1100000.0, 2.001647e+09, 'Titanic')
plt.xscale('log')
plt.show()
```
