```python
# Import the required libraries for the analysis

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
from matplotlib.pyplot import figure
plt.style.use('ggplot')
import os

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)
```

```python
# Located the working directory of Jupyter Notebook

Working_directory = os.getcwd()
print(Working_directory)

# Creat a path and read in the file

path = Working_directory + '/python/movies.csv'
df = pd.read_csv(path)
```

```python
# Explore the top six rows of the data

df.head()

```

```python
# Order by gross in desecending order
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
df.genre.value_counts()
```

```python
# Exploring the data
print(df.shape)
df.describe()
```

```python
# Correlation marix and heatmap

cor = df.corr(method = 'pearson')

# The heatmap shows gross and votes are strongly correlated with budget
sns.heatmap(cor, annot = True)
```

```python
# Display high correlated variables > 0.6
cor = df.corr(method = 'pearson')
Pair = cor
corr_pair = Pair.unstack()
corr_sort = corr_pair.sort_values()
High_corr = corr_sort[(corr_sort) > 0.6]
High_corr
```

```python
# Make a scatter plot of the correlated variables (budget vs gross)

plt.scatter('budget', 'gross', data = df, alpha = 0.9)
plt.title('Budget vs Gross earning')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.text(237000000.0, 2.847246e+09, 'Avatar')
plt.text(356000000.0, 2.797501e+09, 'Avengers: Endgame')
plt.text(200000000.0, 2.201647e+09, 'Titanic')
plt.xscale('log')
plt.show()
```

```python
# Make a scatter plot of the correlated variables (votes vs gross)

plt.scatter('votes', 'gross', data = df, alpha = 0.9 )
plt.title('Budget vs Gross earning')
plt.xlabel('Votes received')
plt.ylabel('Gross Earnings')
plt.text(1100000.0, 2.847246e+09, 'Avatar')
plt.text(903000.0, 2.797501e+09, 'Avengers: Endgame')
plt.text(1100000.0, 2.201647e+09, 'Titanic')
plt.xscale('log')
plt.grid(False)
plt.show()
```

```python
# Result shows that (1) gross earnings is positively associated with cost of production and public perception of the movie
# (2) Avatar followed by Avengers:Endagame and Titanic have the highest earnings.
```
